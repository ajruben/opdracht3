Recognizing the Growing Pattern: Examining the Increase in Divorce Rates in the Netherlands

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv("37296ned_UntypedDataSet_26032024_091942.csv", header = 0, sep=';')
df['Perioden'] = pd.to_datetime(df['Perioden'].str[:4], format='%Y')
```

```python
for i in df.columns:
    print(i)
```

```python
df.set_index('Perioden', inplace=True)
for column in df.columns:
    plt.figure(figsize=(10, 6))
    plt.plot(df.index, df[column], marker='o', linestyle='-', label=column)
    plt.title(f'{column} Over Time')
    plt.xlabel('Year')
    plt.ylabel(column)
    plt.legend()
    plt.grid(True)
    plt.show()
```
